```python
"""Vamos a identificar las columnas duplicadas"""
def encontrar_columnas_duplicadas (cattle_data):
    columnas_duplicadas = cattle_data.columns[cattle_data.columns.duplicated()].unique()
    return columnas_duplicadas
encontrar_columnas_duplicadas (cattle_data)
```

```python
"""Vamos a crear una función que nos elimine las columnas duplicadas. Lo he intentado previamente con drop_duplicates pero en este caso me elimina todas las columnas que tengan duplicados y yo lo que quiero es eliminar todas excepto una de ellas, es decir dejar todas las columnas únicas"""

def eliminar_columnas_duplicadas(df):
    """
    Elimina las columnas duplicadas

    Args:
        df (_pd.Dataframe): El dataframe de entrada
    Returns:
        pd.DataFrame: El dataframe sin columnas duplicadas
    """
    if df.columns.duplicated().any():
        df = df.loc[:, ~df.columns.duplicated()]
    return df
cattle_data_sin_duplicados = eliminar_columnas_duplicadas(cattle_data)
cattle_data_sin_duplicados
```
